```python
from dataclasses import dataclass
from typing import Any
import operator


@dataclass
class Filter:
    """Filter for a DataFrame."""

    key: str
    op: str
    value: Any

    @staticmethod
    def str_to_op(op: str) -> callable:
        """Convert a string to a function."""
        return {
            "==": operator.eq,
            "!=": operator.ne,
            ">": operator.gt,
            "<": operator.lt,
            ">=": operator.ge,
            "<=": operator.le,
            "in": operator.contains,
            "not in": lambda x, y: not operator.contains(x, y),
        }[op]

    def as_tuple(self) -> tuple[str, callable, Any]:
        """Return the filter as a tuple."""
        return self.key, Filter.str_to_op(self.op), self.value
```
